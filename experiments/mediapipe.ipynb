```python
#!/usr/bin/env python3

import                             cv2
import numpy as                    np
import matplotlib.pyplot as        plt
import                             time
import mediapipe as                mp
from mediapipe.tasks import        python
from mediapipe.tasks.python import vision
```

```python
video_path     = "../../data/original_quality/2023-09-11/anit_seated_leg_curl.mp4"
captured       = cv2.VideoCapture(video_path)
video_title    = "Anit seated leg curls"
_, first_frame = captured.read()
first_frame    = cv2.cvtColor(first_frame, cv2.COLOR_BGR2RGB)

plt.title(video_title)

plt.imshow(first_frame)
plt.axis("off")
plt.show()
```

```python
# Initialize the timer
start_time = time.time()

cv2.startWindowThread()

# Main loop for capturing and displaying frames
seconds_to_play = 5
while True:
    # Capture a frame from the video source
    success_flag, frame = captured.read()

    # Calculate frames per second (FPS)
    current_time = time.time()
    elapsed_time = current_time - start_time
    fps = str(int(1 / elapsed_time))
    start_time = current_time  # Update the start time

    # Define text properties for FPS display
    text_location = (70, 50)
    font = cv2.FONT_HERSHEY_PLAIN
    font_weight = 3
    font_color = (255, 0, 0)

    # Add FPS text to the frame
    cv2.putText(frame, fps, text_location, font, font_weight, font_color, font_weight)

    # Display the frame in a window with the specified title
    cv2.imshow(video_title, frame)

    # Check for user input and exit loop if needed
    key = cv2.waitKey(1)
    if (key == 27) or (elapsed_time >= seconds_to_play):  # Exit on 'Esc' key (ASCII code 27)
        break

# Release video capture and close any open windows
cv2.destroyAllWindows()
captured.release()
```
